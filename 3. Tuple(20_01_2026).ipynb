```python
# creating and manipulating tuple.

immutable = ('happy', 'birthday', 'to', 'you')
print(immutable)
```

```python
# accesing using index values:
print(immutable[0])
print(immutable[1])
print(immutable[2])
print(immutable[3])
```

```python
#slicing funtion in tuple
print(immutable[-2])
print(immutable[0:4])
print(immutable[-4:-1])
print(immutable[:-1])
print(immutable[:3])
print(immutable[1:2])
```

```python
for i in immutable:
  print(i)
```

```python
mutable = list(immutable)
print(mutable)
mutable[-1] = "Mohan"
print(mutable)
mutable.index("Mohan")
mutable.insert(3,"Rohan")
print(mutable)
```

```python
for i in mutable:
  print(i)
```
