```python
stack=[]

top = -1
size = 5
choise = 0

def push():
  global top
  global size
  if(top==size):
    print("Stack overflow")

  else:
    top += 1
    data = input("enter your data : ")
    stack.append(data)

def pop():
  global top
  if(top==-1):
    print("stack underflow")

  else:
    top=top-1
    stack.pop()

while(choise!=1):
  print("Your Choises: \n 1.PUSH \n 2.POP \n 3.TOP \n 4.SIZE \n 5.EXIT \n ")
  ch = int(input("Enter your choise : "))

  match(ch):
    case 1:
      push()

    case 2:
      pop()

    case 3:
      if(top==-1):
        print("Stack is empty")
      else:
        print(f"Top value is : {stack[top]}")

    case 4:
      print(f"Size of the stack is : {size}")

    case 5:
      print("Ending Program ...")
      choise +=1

    case _:
      print ("Wrong Choise !!")


```
