```python
#1. Check whether a number is even or odd

num = int(input("Enter a number: "))

if(num==0):
  print(f"The number is {num} which is nither odd or even")

elif(num%2==0):
  print(f"The number {num} is even")

else:
  print(f"The given number {num} is odd")
```

```python

#4. find the largest of three numbers using nested-if
a = int(input("Enter the first number"))
b = int(input("Enter the second number"))
c = int(input("Enter the third number"))

if(a>b):
  if(a>c):
    print(f"{a} is greater than {b} and {c}")
  else:
    print(f"{a} is greater than {b} but not {c}")

elif(b>a):
  if(b>c):
    print(f"{b} is greater than {a} and {c}")
  else:
    print(f"{b} is greater than {a} but not {c}")

else:
  if(c>a):
    print(f"{c} is greater than {a} and {b}")
  else:
    print(f"{c} is greater than {a} but not {b}")




```

```python
# 7.	Find the factorial of a number
num = int(input("Enter a number : "))
result=1
for i in range(1,num+1):
  result *= i
print(f"The Factorial of {num} is : {result}")
```

```python
# 8.	Check whether a number is prime
num = int(input("Enter a number : "))
flag = 0
for i in range(2,num):
  if(num%i == 0):
    flag+=1
if(flag==0):
  print(f"The number {num} is prime")
else:
  print(f"The number {num} is not prime")
```
