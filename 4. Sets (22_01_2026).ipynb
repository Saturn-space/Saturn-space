```python
alphabet = {"A","B","C"}
numbers = {1,2,3,4}

alphabet.add("D")
print(alphabet)

alphabet.update(numbers)
print(alphabet)

alphabet.remove("D")
print(alphabet)

alphabet.pop()
print(alphabet)

#alphabet.clear()
del alphabet

alphabet = {"A","B","C"}
print(alphabet)

```

```python
# playting with dictionary

objects = {
    "A":"Apple",
    "B":"Ball",
    "C":"Cat",
    "D":"Dog"
}

print(objects)
print(objects)
print(objects["A"])

obj2 = {
    "size" :(20,30,40),
    "brand" : ("A1","B2","C3")
}
print(obj2)
print(obj2["size"])

```

```python
objects.keys()
objects.values()
objects.items()

print(objects.keys())
print(objects.values())
print(objects.items())
```

```python
objects["A"] = "Ant"
print(objects)

objects.update(obj2)
print(objects)

obj2.update({"B":"Bat"})
print(obj2)

objects.update({"B":"Bat"})
print(objects)
```

```python
for x in objects:
  print(x)

for x in objects.keys():
  print(x)

for x in objects.values():
  print(x)

for x,y in objects.items():
  print(x,y)

print(objects)
```
