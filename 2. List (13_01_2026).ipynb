```python
animals = ["Tiger","Lion","Elephant","Zebra","Girraf"]

# printing
print(animals[0])
print(animals[1:4])
print(animals[:4])
print(animals[3:])
print(animals[1:-2])
print(animals[-1:-2])
print(animals[-2])
print(animals[-1:-4]) # it will return an empty list because the printing or elements are accessed by the machine start from left to right ;
print(animals[-4:-1])

# Insert
animals.insert(1,"Crow")
animals.insert(2,"Hippo")
print(animals)

# Append
animals.append("Monkey")
animals.append("Ostritch")
print(animals)

#length of the list
print(len(animals))

#removing
print(animals.pop(1))
animals.pop(-1)
print(animals)






```

```python
Students = []
for i in range(5):
  name = input("Enter the name of the student: ")
  Students.append(name)
print(Students)

Students.pop()  #it delete the last element if index is not known.
print(Students)

Students.pop(0)  # it delete the element whose index is specified.
print(Students)

Students.remove("rohan")  #it delete the element mention in it if there are duplicates value then the first occurance of the value in the list is deleted.
print(Students)
```
