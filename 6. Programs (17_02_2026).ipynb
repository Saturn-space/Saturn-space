```python
#1. program to varify single digit exist in a list if more than single exits add them and print the new list.

number = [22,4,53,87,34,2]

def sum(arg):
  new_number = []

  for i in arg:
      if(i>=10 and i<=99):
        value = 0;
        while i!=0:
          value += i%10
        new_number.append(value)

      else:
        new_number.append(i)

  return new_number

# new_num = sum(number)
# print(new_num)
new_number = []
for i in number:
      if(i>=10 and i<=99):
        value = 0;
        while i!=0:
          value += i%10
          i = i//10;
        new_number.append(value)
      else:
        new_number.append(i)

print(new_number)
```

```python
#2. Program to add items along with some details and find the total payable amount.

stock = {}

def items():
  n = int(input("Enter the number of items :  "))

  for i in range(n):
    id = int(input("Enter the Product ID : "))
    name = input("Enter the Product Name : ")
    price = int(input("Enter the Product Price : "))
    Qty = int(input("Enter the Product Quntity : "))
    Brand = input("Enter the Product Brand : ")

    stock[id] = [name,price,Qty,Brand]

  print(stock)

def total():
  sum = 0
  for i in stock.values():
    sum += i[1]*i[2]

  print(f"Total Payable Amount : ${sum}")

items()
total()




```
