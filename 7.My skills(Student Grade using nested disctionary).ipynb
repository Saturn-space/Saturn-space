```python
#PROGRAM TO FIND THE GRADE OF THE STUDENT (IN MY STYLE):
student = {
    "student1":{
        "name":"Raj",
        "roll":"100",
        "subject":{
            "English":89,
            "hindi":39,
            "maths":67,
            "science":93,
            "social science":66
        }
    },
    "student2":{
        "name":"sam",
        "roll":200,
        "subject":{
            "English":49,
            "hindi":78,
            "maths":56,
            "science":73,
            "social science":45
        }
    },
    "student3":{
        "name":"Tony",
        "roll":300,
        "subject":{
            "English":59,
            "hindi":89,
            "maths":57,
            "science":83,
            "social science":62
        }
    },
    "student4":{
        "name":"Happy",
        "roll":400,
        "subject":{
            "English":89,
            "hindi":79,
            "maths":51,
            "science":82,
            "social science":56
        }
    },
    "student5":{
        "name":"Ali",
        "roll":500,
        "subject":{
            "English":79,
            "hindi":45,
            "maths":99,
            "science":63,
            "social science":82
        }
    },
    "student6":{
        "name":"shyam",
        "roll":600,
        "subject":{
            "English":64,
            "hindi":49,
            "maths":75,
            "science":75,
            "social science":80
        }
    },
    "student7":{
        "name":"Btech",
        "roll":700,
        "subject":{
            "English":0,
            "hindi":85,
            "maths":74,
            "science":58,
            "social science":70
        }
    },
    "student8":{
        "name":"Sony",
        "roll":800,
        "subject":{
            "English":59,
            "hindi":89,
            "maths":57,
            "science":83,
            "social science":62
        }
    }
}

def avg(x):
    sum=0
    sub=0
    for i in student[f"{x}"]["subject"]:
        sum += student[x]["subject"][i]
        sub+=1
        result=sum/sub
    return result

def grade(x):
    value = 0
    if(x>=90):
        value = "A"
    elif(x>=80):
        value = "B"
    elif(x>=70):
        value = "C"
    elif(x>=60):
        value = "D"
    else:
        value = "F"

    return value

def show_result():
    global all
    global num
    all=0
    num=0
    for i in student:
        save = avg(i)
        result = grade(save)
        print(f"""
        Name: {student[i]["name"]}
        Roll No: {student[i]["roll"]}
        Percentage: {save}
        Grade: {result}
        """)
        # global all
        num += 1
        all += save

def show_performance(x,y):
    perf = x/y
    print(f"Overall class Performance : {perf} ")


show_result()
show_performance(all,num)


```
